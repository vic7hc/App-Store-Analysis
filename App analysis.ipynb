```python
import pandas as pd #data analysis
import matplotlib.pyplot as plt #for plotting
import seaborn as sns
df = pd.read_csv('AppleStore.csv')
df.head()
```

```python
corr_df = df.corr()
corr_df
```

```python
len(df)
```

```python
print("The free apps are:", len(df[df.price == 0]) )
print("The paid apps are:", len(df[df.price > 0]) )
print("The most expensive apps are:", len(df[df.price > 50]) )
print("So, we gonna remove the free and the most expensive:")
expensive = df[df.price > 50]
expensive
```

```python
range_apps = df[((df.price < 50) & (df.price > 0))]
print ("So, the max price will be: ", max(range_apps.price))
print ("And the min price will be: ",  min(range_apps.price))
```

```python
prime_genre = df["prime_genre"].unique().tolist()
prime_genre.sort()
add_prime = len(prime_genre)
add_prime
```

```python
plt.style.use('fivethirtyeight')
plt.figure(figsize=(10,10))
plt.hist(range_apps.price,log=True)
plt.show()
```

```python
plt.figure(figsize=(15, 2))
plt.title('Visual price distribution')

yrange = [0,25]
plt.xlim(yrange)
sns.stripplot(data=range_apps,y = 'price',jitter= True,orient = 'h' ,size=6)
plt.show()
```

```python
plt.figure(figsize=(15, 10))

plt.subplot(4,1,1)
plt.title('Games')
yrange = [0,25]
plt.xlim(yrange)
games = range_apps[range_apps.prime_genre == 'Games']
sns.stripplot(data = games, y='price',jitter = True , orient = 'h', size = 6, color ='#eb5e66')

plt.subplot(4,1,2)
plt.title('Education')
yrange = [0,25]
plt.xlim(yrange)
games = range_apps[range_apps.prime_genre == 'Education']
sns.stripplot(data = games, y='price',jitter = True , orient = 'h', size = 6, color ='blue')

plt.subplot(4,1,3)
plt.title('Entertainment')
yrange = [0,25]
plt.xlim(yrange)
games = range_apps[range_apps.prime_genre == 'Entertainment']
sns.stripplot(data = games, y='price',jitter = True , orient = 'h', size = 6, color ='black')
```
